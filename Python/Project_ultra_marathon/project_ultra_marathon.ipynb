```python
# questões que podem ser respondidas a partir dos dados 
```

```python
#Qual a média de idade dos atletas?
average_age = df3['athete_age'].mean()
average_age
```

```python
#Diferença de velocidade entre homens e mulheres nas distâncias de 50km e 50mi?
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#Quais faixas etárias são as melhores na corrida de 50mi?
df3.query('race_length == "50mi"').groupby('athete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False).query('count >= 20')
```

```python
#Quais faixas etárias são as piores na corrida de 50mi?
df3.query('race_length == "50mi"').groupby('athete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=True).query('count > 9')
```

```python
#A velocidade média dos atletas é mais lento no verão do que no inverno?

#Primavera (spring) 3-5
#Verão (summer) 6-8
#Outono (fall)9-11
#Inverno (winter)12 - 2
```

```python
df3['race_month']= df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'spring' if x in [3,4,5] else 'summer' if x in [6,7,8] else 'fall' if x in [9,10,11] else 'winter')
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
```
