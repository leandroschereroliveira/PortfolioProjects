**Analise de dados**

**analise univariada ou univariate analysis**<br>
Consiste basicamente em, para cada uma das variáveis individualmente: Classificar a variável quanto a seu tipo: qualitativa (nominal ou ordinal) ou quantitativa (discreta ou contínua)

```python
#Price distribution
data['price'].describe()
```

```python
sns.histplot(data['price'], bins=50, kde=True)
```

```python
#Price distribution without outliers
sns.boxplot(data= data, x='price')
```

```python
#removendo outliers
df = data[data['price'] < 1500]
```

```python
sns.boxplot(data= df, x='price')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data =df ,x='availability_365')
plt.title('Availability_365 distribution')
plt.ylabel('Frequency')
plt.xlabel('Availability_365')
plt.show()
```

```python
#Neighbourhood distribution
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
#Engenharia de recursos
#['price per bed']
df['price_per_bed'] = df['price'] / df['beds']
```

```python
df.head()
```

```python
#average price per bed
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

**analise bivariada ou bi variable analysis**<br>
 Uma das formas mais simples de análise quantitativa. Envolve a análise de duas variáveis, com o objetivo de determinar a relação empírica entre elas. A análise bivariada pode ser útil para testar hipóteses simples de associação

```python
df.columns
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.title('Price by Neighbourhood and Room Type')
plt.ylabel('Price')
plt.xlabel('Neighbourhood')
plt.show()
```

```python
#'number_of_reviews' x 'price'
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.title('Number of Reviews x Price')
plt.ylabel('Price')
plt.xlabel('Number of Reviews')
plt.show()
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
#geographical distribution
plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical distribution of listings')
plt.show()
```

```python
#heatmap - correlation of one variable with others for numerical columns
corr = df[['latitude','longitude','price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data=corr, annot=True)
plt.title('Correlation heatmap')
plt.show()
```
